```python
import pickle
import matplotlib.pyplot as plt

with open('experimentResults/reconstructStobbe2.pkl', 'rb') as f:
        n = pickle.load(f)
        no_edges = pickle.load(f)
        reconErr = pickle.load(f)
#print(reconErr)
```

```python
plot_x_stob = {}
plot_y_stob = {}
for key in reconErr:
    filenum, C = key
    error , sampCplx = reconErr[key]
    try:
        plot_x_stob[filenum].append(sampCplx)
        plot_y_stob[filenum].append(error/no_edges)
    except KeyError:
        plot_x_stob[filenum]=[]
        plot_x_stob[filenum].append(sampCplx)
        plot_y_stob[filenum] =[]
        plot_y_stob[filenum].append(error/no_edges)
#print(plot_x_stob, plot_y_stob)
```
